```python
from queue import PriorityQueue

class Graph:
    def __init__(self):
        self.graph = {
    "/": [(0, ("/","A"))],
"A": [(4, ("A", "B")), (3, ("A", "C"))],
"B": [(4, ("B", "A")), (12, ("B", "E")),(5, ("B", "F"))],
"C": [(3, ("C", "A")), (7, ("C", "D")),(10, ("C", "E"))],
"D": [(7, ("D", "C")), (2, ("D", "E"))],
"E": [(12, ("E", "B")), (10, ("E", "C")), (5, ("E", "Z")),(2, ("E", "D"))],
"F": [(5, ("F", "B")), (16, ("F", "Z"))],
"Z": [(5, ("Z", "E")), (16, ("Z", "F"))]
                            }
        self.edges = {}
        self.weights = {}
        self.heristics = {
            "A" : 14,
            "B" : 12,
            "C" : 11,
            "D" : 6,
            "E" : 4,
            "F" : 11,
            "Z": 0
        }
        self.populate_edges()
        self.populate_weights()

    def populate_edges(self):
        for key in self.graph:
            neighbours = []
            for each_tuple in self.graph[key]:
                neighbours.append(each_tuple[1][1])
            self.edges[key] = neighbours

    def populate_weights(self):
        for key in self.graph:
            neighbours = self.graph[key]
            for each_tuple in neighbours:
                self.weights[each_tuple[1]] = each_tuple[0]

    def neighbors(self, node):
        return self.edges[node]

    def get_cost(self, from_node, to_node):
        return self.weights[(from_node,  to_node)]

    def get_heuristic(self, node):
        return self.heristics[node]

def reconstruct_path(came_from, current):
    total_path = [current]
    while current in came_from:
        current = came_from[current]
        total_path.append(current)
    return total_path[::-1]

def astar_with_path(graph, start, goal):
    open_set = PriorityQueue()
    open_set.put((graph.get_heuristic(start), start))

    came_from = {}
    g_score = {node: float('inf') for node in graph.edges}
    g_score[start] = 0


    f_score = {node: float('inf') for node in graph.edges}
    f_score[start] = graph.get_heuristic(start)

    while not open_set.empty():
        current_f_score, current_node = open_set.get()

        if current_node == goal:
            path = reconstruct_path(came_from, current_node)
            return path, g_score[goal]

        for neighbor in graph.neighbors(current_node):
            tentative_g_score = g_score[current_node] + graph.get_cost(current_node, neighbor)

            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current_node
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + graph.get_heuristic(neighbor)
                open_set.put((f_score[neighbor], neighbor))

    return [], float('inf')


graph_instance = Graph()
path, cost = astar_with_path(graph_instance, "A", "Z")
print("Optimal Path : ", path)
print("Total Cost : ", cost)
```
